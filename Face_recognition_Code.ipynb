# Resim İşlemleri

```python
# Gerekli kütüphaneleri ekliyoruz.
import cv2
import face_recognition
# Bu kodun hangi kişiyi tanımasını istiyorsak o kişiye ait bir resmi yüklüyoruz.
# Burada o kişinin yüzü net çıkmalı ve resimde başka kimse olmamalıdır.
# Ben burada bu kod Messi'yi tanısın diye Messi'ye ait bir resmi yükledim.
main_image=face_recognition.load_image_file("messi.jpg") 
# Hangi kişiyi tanımasını istiyorsak onun adını yazıyoruz.
person= 'Messi'
# Burada resimdeki kişinin yüzüne ait bir encoding(kodlama) oluşturuyor.
# Burada ilk indisi alıyoruz. Ama yüklediğimiz resimde tek bir kişi
# olduğu için zaten  sadece bir indis olacaktır.
encoding1=face_recognition.face_encodings(main_image)[0] 
# Burada diğer resmi yüklüyoruz. Yüz tanıma işlemleri bu resimde yapılacak.
img=cv2.imread("messi2.jpg")
# Resimdeki yüzleri tespit edip resimdeki konumları buluyoruz.
faceloc=face_recognition.face_locations(img)
# Test resmindeki yüzler için bir encoding oluştuyoruz. 
# Test resminde yaparken yüzlerin konumlarını da kullanıyoruz.
encoding2=face_recognition.face_encodings(img,faceloc)
# Burada iki encoding karşılaştırılıp yüzlerin aynı kişiye ait olup olmadığı tespit ediliyor.
# Burada bir liste dönüyor. Bu listede test resmindeki her bir kişi için bir boolean ifade var.
# Eğer buradaki kişiler aynıysa True değilse False vardır bu liste içinde.
matchedfaces=face_recognition.compare_faces(encoding1,encoding2)
# Eğer True varsa yani bir eşleşme varsa bu bloğun içine girilir.
if True in matchedfaces:
    # Burada yüz konumlarını kullanarak kişilerin yüzünü kare içine alıyoruz.
    # Ayrıca yüzü eşleştiği içini adını yazdırıyoruz.
    for (y,x2,y2,x) in faceloc:
        cv2.rectangle(img,(x,y),(x2,y2),(255,0,0),3)
        cv2.putText(img, person,(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2, (150, 0, 200), 2)
else:
    for (y,x2,y2,x) in faceloc:
        cv2.rectangle(img,(x,y),(x2,y2),(255,0,0),3)
        # Burada yüzü eşleşmediği tanımlanamayan insan yazdırıyoruz.
        cv2.putText(img, "Undefined",(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2, (150, 0, 200), 2)        
# Resmi ekranda gösteriyoruz.       
cv2.imshow("Resim",img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
# Burada test resminde tek kişi olduğu varsayılmıştır. 
# Birden fazla kişinin olduğu durumlar için diğer örneğe bakınız.
```

# Video İşlemleri

Burada yüz bulma ve yüze ait encoding oluşturmak uzun süren bir işlem. Bu yüzden videolarda FPS çok düşük oluyor.

```python
import cv2
import face_recognition

main_image=face_recognition.load_image_file("messi.jpg") 
person= 'Messi'
encoding1=face_recognition.face_encodings(main_image)[0] 
# Burada kendi videonuzun adını yazıyorsunuz.
kamera= cv2.VideoCapture('video5.mp4')
while True:
    ret,img=kamera.read()
    faceloc=face_recognition.face_locations(img)
    encoding2=face_recognition.face_encodings(img,faceloc)
    # Video açıldığında videoda bir kişi yoksa encoding2 boş liste
    # olur ve hataya sebep olur. O yüzden videoda bir kişi tespit edince 
    # kalan işlemlere devam edilecek
    if encoding2!=[]:
        matchedfaces=face_recognition.compare_faces(encoding1,encoding2)
        a=0
        for i in matchedfaces:
            if i:
                y,x2,y2,x=faceloc[a]
                cv2.rectangle(img,(x,y),(x2,y2),(0,255,0),2)
                cv2.putText(img, person,(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2, (150, 0, 200), 2)   
            else:
                y,x2,y2,x=faceloc[a]
                cv2.rectangle(img,(x,y),(x2,y2),(0,255,0),2)
                cv2.putText(img, "Undefined",(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2, (150, 0, 200), 2) 
            a=a+1

    cv2.imshow("kamera",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
kamera.release()
cv2.destroyAllWindows()
```

# Birden Fazla Kişi Tanıyan Kod

Burada ilk önce hangi kişileri tanımasını istiyorsak bu kişilerin resmini yükleyip bir listeye kaydediyoruz.

```python
import cv2
import face_recognition
# Burada yüzlerin encoding'lerini bu listeye ekleyeceğiz.
encoding_list=[]
# Buraya da kişilerin adlarını ekleyeceğiz.
isimler=[]

main_image1=face_recognition.load_image_file("messi.jpg") 
person1= 'Messi'
encoding1=face_recognition.face_encodings(main_image1)[0] 
encoding_list.append(encoding1)
isimler.append(person1)

main_image2=face_recognition.load_image_file("ronaldo.jpg") 
person2= 'Ronaldo'
encoding2=face_recognition.face_encodings(main_image2)[0] 
encoding_list.append(encoding2)
isimler.append(person2)


main_image3=face_recognition.load_image_file("neymar.jpg") 
person3= 'Neymar'
encoding3=face_recognition.face_encodings(main_image3)[0] 
encoding_list.append(encoding3)
isimler.append(person3)


main_image4=face_recognition.load_image_file("mbappe.jpg") 
person4= 'Mbappe'
encoding4=face_recognition.face_encodings(main_image4)[0] 
encoding_list.append(encoding4)
isimler.append(person4)


main_image5=face_recognition.load_image_file("bale.jpg") 
person5= 'Bale'
encoding5=face_recognition.face_encodings(main_image5)[0] 
encoding_list.append(encoding5)
isimler.append(person5)

```

Burada yukarıdaki kodlardan farklı olarak resimdeki yüzleri bulmak için bir yolov8 modeli kullanacağız. Böylece kod daha hızlı çalışıyor. Siz isterseniz yukarıdaki yöntemi kullanabilirsiniz. Modeli GitHub'a yükledim oradan bulabilirsiniz.

```python
from ultralytics import YOLO
model= YOLO("yolov8n-face.pt")
```

Buradaki fonksiyon ile yüzlerin konumunu buluyoruz.

```python
def tespitet(img):
    location=[]

    
    rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    results = model(rgb_img,verbose=False) 
    
    for i in range(len(results[0].boxes)):
        x1,y1,x2,y2=results[0].boxes.xyxy[i]
        score=results[0].boxes.conf[i]
     
    
        x1,y1,x2,y2,score=int(x1),int(y1),int(x2),int(y2),float(score)
        if score<0.5:
            continue
            
  
      
            

        location.append((y1,x2,y2,x1))
    return location
```

```python
img=cv2.imread("mnm.jpg")
faceloc=tespitet(img)
encoding=face_recognition.face_encodings(img,faceloc)
# bu listede resimde kaç tane yüz varsa o kadar eleman var.
# Her bir eleman bir yüzü temsil edecek.
# Bu liste sayesinde resimde eşleşmeyen yüzleri bulup undefined yazdıracağız.
# Eğer resimdeki bir yüz veri tabanındaki bir yüz eşleşmişse bu listeden çıkarılacak.

location_list= [x for x in range(len(faceloc))]
if encoding!=[]:
    
    
    # Burada veri tabanındaki her bir yüzün encoding'i ile resimdeki yüzlerin encoding'lerini kıyaslıyoruz.
    for indis,enc in enumerate(encoding_list):
        

        matchedfaces=face_recognition.compare_faces(enc,encoding)
        a=0
        
        for i in matchedfaces:
            # Burada sadece eşleşme varsa ekrana yazdırıyoruz.
            if i:
                # Eşleşme olan yüzü listeden çıkarıyoruz.
                location_list.remove(a)
                
                y,x2,y2,x=faceloc[a]
                cv2.rectangle(img,(x,y),(x2,y2),(0,255,0),3)
                # Burada eşleşen yüzün hangi kişiye ait olduğunu buluyoruz.
                person=isimler[indis]
                cv2.putText(img, person,(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2.2, (102, 0, 153), 2)  
            
            
            a=a+1
            
            
    # Tüm eşleşme işlemleri bittikten sonra listede kalan ve eşleşmeyen bu yüzleri undefined diye gösteriyoruz.
    for a in location_list:
        
        y,x2,y2,x=faceloc[a]
        cv2.rectangle(img,(x,y),(x2,y2),(0,255,0),5)
        cv2.putText(img, "Undefined",(x,y-10), cv2.FONT_HERSHEY_COMPLEX, 2.2, (255, 0, 255), 2) 
        
       

cv2.imshow("Resim",img) 


cv2.waitKey(0)
cv2.destroyAllWindows()
```
